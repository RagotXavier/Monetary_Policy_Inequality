```julia
function describe_simulation(calib::String, num_files::Int) 
    
    df_moments = DataFrame()
    df_corr    = DataFrame()


    if num_files ≥ 1
        file1 = matopen("../dynamics/moments_eco1_$(calib).mat")
        moments1 = read(file1)

        df_moments = DataFrame(Variable=vec(moments1["list_table"]), Mean_Eco_1=vec(moments1["mean_var_table"]), Std_Eco_1=vec(moments1["std_norm_table"]))
        df_corr = DataFrame(Variable=vec(moments1["corr_list"]), corr_Eco_1=vec(moments1["corr"]))

        close(file1)  
    end


    if num_files ≥ 2
        file2 = matopen("../dynamics/moments_eco2_$(calib).mat")
        moments2 = read(file2)

        df_moments.Mean_Eco_2 = vec(moments2["mean_var_table"])
        df_moments.Std_Eco_2 = vec(moments2["std_norm_table"])
        df_corr.corr_Eco_2 = vec(moments2["corr"])

        close(file2) 
    end


    if num_files ≥ 3
        file3 = matopen("../dynamics/moments_taylor.mat")
        moments3 = read(file3)

        df_moments.Mean_Eco_3 = vec(moments3["mean_var_table"])
        df_moments.Std_Eco_3 = vec(moments3["std_norm_table"])
        df_corr.corr_Eco_3 = vec(moments3["corr"])

        close(file3) 
    end

    if num_files ≥ 4
        file4 = matopen("../dynamics/moments_eco3_$(calib).mat")
        moments4 = read(file4)

        df_moments.Mean_Eco_4 = vec(moments4["mean_var_table"])
        df_moments.Std_Eco_4 = vec(moments4["std_norm_table"])
        df_corr.corr_Eco_4 = vec(moments4["corr"])

        close(file4) 
    end



    # file1 = matopen("../dynamics/moments_eco1_$(calib).mat")
    # file2 = matopen("../dynamics/moments_eco2_$(calib).mat")
    # file3 = matopen("../dynamics/moments_eco3_$(calib).mat")

    # moments1 = read(file1) 
    # moments2 = read(file2) 
    # moments3 = read(file3) 

    # # df_moments = DataFrame(Variable=vec(moments["list"]), Mean=vec(moments["mean"]), Std=vec(moments["std_norm"]))
    # # df_corr    = DataFrame(Variable=vec(moments["corr_list"]), corr=vec(moments["corr"]))

    # df_moments = DataFrame(Variable=vec(moments1["list"]), Mean_Eco_1=vec(moments1["mean"]), Std_Eco_1=vec(moments1["std_norm"]), Mean_Eco_2=vec(moments2["mean"]), Std_Eco_2=vec(moments2["std_norm"]),Mean_Eco_3=vec(moments3["mean"]), Std_Eco_3=vec(moments3["std_norm"]))
    # df_corr    = DataFrame(Variable=vec(moments1["corr_list"]), corr_Eco_1=vec(moments1["corr"]),corr_Eco_2=vec(moments2["corr"]),corr_Eco_3=vec(moments2["corr"]))


    display(df_moments)
    display(df_corr)
    return 
end
```
